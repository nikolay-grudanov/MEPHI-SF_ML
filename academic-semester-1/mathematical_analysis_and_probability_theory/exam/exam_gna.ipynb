# Решение 

## а) 
$$y=e^{x \ln (3 x+1)}$$

Шаг 1: Применение правила дифференцирования сложной функции

Функция имеет вид $y = e^u$, где $u = x \ln (3x+1)$. Применим правило дифференцирования сложной функции:

$$\frac{dy}{dx} = e^u \cdot \frac{du}{dx}$$

Шаг 2: Вычисление $\frac{du}{dx}$

Теперь нужно найти производную $\frac{du}{dx}$. Здесь $u$ является произведением двух функций: $x$ и $\ln (3x+1)$. Применим правило дифференцирования произведения:

$$\frac{du}{dx} = 1 \cdot \ln (3x+1) + x \cdot \frac{d}{dx}[\ln (3x+1)]$$

Шаг 3: Вычисление производной логарифма

Для нахождения $\frac{d}{dx}[\ln (3x+1)]$ применим правило дифференцирования логарифма:

$$\frac{d}{dx}[\ln (3x+1)] = \frac{1}{3x+1} \cdot \frac{d}{dx}[3x+1] = \frac{1}{3x+1} \cdot 3 = \frac{3}{3x+1}$$

Шаг 4: Подстановка результата в выражение для $\frac{du}{dx}$

Подставим полученное выражение обратно в формулу для $\frac{du}{dx}$:

$$\frac{du}{dx} = \ln (3x+1) + x \cdot \frac{3}{3x+1}$$

Шаг 5: Окончательное выражение производной

Теперь подставим полученное выражение для $\frac{du}{dx}$ в формулу из шага 1:

$\frac{dy}{dx} = e^{x \ln (3x+1)} \cdot \left(\ln (3x+1) + x \cdot \frac{3}{3x+1}\right)$

Итоговый результат

Таким образом, производная функции $y = e^{x \ln (3x+1)}$ имеет вид:

$$\frac{dy}{dx} = e^{x \ln (3x+1)} \cdot \left(\ln (3x+1) + \frac{3x}{3x+1}\right)$$

Это и есть окончательный ответ задачи.

**Ответ:** $\frac{dy}{dx} = e^{x \ln (3x+1)} \cdot \left(\ln (3x+1) + \frac{3x}{3x+1}\right)$

## б)
$$y=e^{x^{2}} \sqrt{x^{3}+4 x^{2}-7}$$

Давайте найдем производную данной функции, используя правила дифференцирования. Разберем решение пошагово.

Дана функция:

$y = e^{x^2} \sqrt{x^3 + 4x^2 - 7}$

Шаг 1: Применение правила дифференцирования произведения

Наша функция представляет собой произведение двух функций: $u = e^{x^2}$ и $v = \sqrt{x^3 + 4x^2 - 7}$. Применим правило дифференцирования произведения:

$$\frac{dy}{dx} = u' \cdot v + u \cdot v'$$

где $u' = \frac{d}{dx}(e^{x^2})$ и $v' = \frac{d}{dx}(\sqrt{x^3 + 4x^2 - 7})$

Шаг 2: Нахождение $u'$

Для нахождения $u'$ применим правило дифференцирования сложной функции:

$$u' = e^{x^2} \cdot \frac{d}{dx}(x^2) = e^{x^2} \cdot 2x$$

Шаг 3: Нахождение $v'$

Для нахождения $v'$ применим правило дифференцирования корня и правило дифференцирования сложной функции:

$$v' = \frac{1}{2\sqrt{x^3 + 4x^2 - 7}} \cdot \frac{d}{dx}(x^3 + 4x^2 - 7)$$

$$\frac{d}{dx}(x^3 + 4x^2 - 7) = 3x^2 + 8x$$

Таким образом:

$$v' = \frac{3x^2 + 8x}{2\sqrt{x^3 + 4x^2 - 7}}$$

Шаг 4: Подстановка в формулу производной произведения

Теперь подставим полученные выражения для $u'$ и $v'$ в формулу производной произведения:


$$\frac{dy}{dx} = (e^{x^2} \cdot 2x) \cdot \sqrt{x^3 + 4x^2 - 7} + e^{x^2} \cdot \frac{3x^2 + 8x}{2\sqrt{x^3 + 4x^2 - 7}}$$


Шаг 5: Упрощение выражения


Приведем полученное выражение к общему знаменателю:

\begin{align}

    \frac{dy}{dx} = e^{x^2} \cdot \frac{4x\sqrt{x^3 + 4x^2 - 7} \cdot \sqrt{x^3 + 4x^2 - 7} + 3x^2 + 8x}{2\sqrt{x^3 + 4x^2 - 7}}

    \\

    \frac{dy}{dx} = e^{x^2} \cdot \frac{4x(x^3 + 4x^2 - 7) + 3x^2 + 8x}{2\sqrt{x^3 + 4x^2 - 7}}

\end{align}


Итоговый результат

Таким образом, производная функции $y = e^{x^2} \sqrt{x^3 + 4x^2 - 7}$ имеет вид:

$$\frac{dy}{dx} = e^{x^2} \cdot \frac{4x^4 + 16x^3 - 28x + 3x^2 + 8x}{2\sqrt{x^3 + 4x^2 - 7}}$$

или после упрощения:

$$\frac{dy}{dx} = e^{x^2} \cdot \frac{4x^4 + 16x^3 + 3x^2 - 20x}{2\sqrt{x^3 + 4x^2 - 7}}$$


**Ответ:** $\frac{dy}{dx} = e^{x^2} \cdot \frac{4x^4 + 16x^3 + 3x^2 - 20x}{2\sqrt{x^3 + 4x^2 - 7}}$

## в)
$$y=\frac{x \sqrt{1-x^{2}}}{1+x^{2}}$$


 Шаг 1: Применение правила дифференцирования частного

Пусть $u = x \sqrt{1-x^{2}}$ и $v = 1+x^{2}$. Тогда наша функция имеет вид $y = \frac{u}{v}$.

Правило дифференцирования частного гласит:

$$\frac{d}{dx}\left(\frac{u}{v}\right) = \frac{v\frac{du}{dx} - u\frac{dv}{dx}}{v^2}$$

 Шаг 2: Нахождение $\frac{du}{dx}$

Для нахождения $\frac{du}{dx}$ применим правило дифференцирования произведения и правило дифференцирования сложной функции:

\begin{align}

    \frac{du}{dx} = \frac{d}{dx}(x \sqrt{1-x^{2}}) = \sqrt{1-x^{2}} + x \cdot \frac{d}{dx}(\sqrt{1-x^{2}})

    \\

    \frac{d}{dx}(\sqrt{1-x^{2}}) = \frac{1}{2\sqrt{1-x^{2}}} \cdot \frac{d}{dx}(1-x^{2}) = \frac{1}{2\sqrt{1-x^{2}}} \cdot (-2x) = -\frac{x}{\sqrt{1-x^{2}}}

\end{align}

Подставляем обратно:

$$\frac{du}{dx} = \sqrt{1-x^{2}} + x \cdot \left(-\frac{x}{\sqrt{1-x^{2}}}\right) = \sqrt{1-x^{2}} - \frac{x^2}{\sqrt{1-x^{2}}}$$

 Шаг 3: Нахождение $\frac{dv}{dx}$

$$\frac{dv}{dx} = \frac{d}{dx}(1+x^{2}) = 2x$$

 Шаг 4: Подстановка в формулу производной частного

Теперь подставим все найденные выражения в формулу производной частного:

$$\frac{dy}{dx} = \frac{(1+x^{2})(\sqrt{1-x^{2}} - \frac{x^2}{\sqrt{1-x^{2}}}) - x\sqrt{1-x^{2}} \cdot 2x}{(1+x^{2})^2}$$

 Шаг 5: Упрощение выражения

 Упростим выражения $\sqrt{1-x^{2}} - \frac{x^2}{\sqrt{1-x^{2}}}$

\begin{align}

    \frac{\sqrt{1-x^{2}}}{1} - \frac{x^2}{\sqrt{1-x^{2}}} = \frac{1-x^{2}}{\sqrt{1-x^{2}}} - \frac{x^2}{\sqrt{1-x^{2}}} = \frac{1-2x^2}{\sqrt{1-x^{2}}}

\end{align}

Полученное значение подставим в наше выражене и раскроем скобки

\begin{align}

    \frac{dy}{dx} = \frac{(1+x^{2})(\frac{1-2x^2}{\sqrt{1-x^{2}}}) - x\sqrt{1-x^{2}} \cdot 2x}{(1+x^{2})^2}
    \\[5em] 
    \frac{dy}{dx} = \frac{\frac{1-2x^2}{\sqrt{1-x^{2}}} + \frac{x^2-2x^4}{\sqrt{1-x^{2}}} - 2x^{2}\sqrt{1-x^{2}}}{(1+x^{2})^2}

\end{align}

Домножим вернию и нижнию часть на $\sqrt{1-x^{2}}$

\begin{align}

    \frac{dy}{dx} = \frac{1 - 2x^2 + x^2 - 2x^4 - 2x^2(1-x^2)}{(1+x^{2})^2 \cdot \sqrt{1-x^{2}}}
    \\[5em] 
    \frac{dy}{dx} = \frac{1 - 2x^2 + x^2 - 2x^4 - 2x^2 + 2x^4}{(1+x^{2})^2 \cdot \sqrt{1-x^{2}}}
    \\[5em] 
    \frac{dy}{dx} = \frac{1 - 3x^2}{(1+x^{2})^2 \cdot \sqrt{1-x^{2}}}

\end{align}

Это и есть окончательный ответ задачи.

**Ответ:** $\frac{dy}{dx} = \frac{1 - 3x^2}{(1+x^{2})^2 \cdot \sqrt{1-x^{2}}}$

# Задание 2

Найдите безусловные экстремумы функции:

$$u=3 x^{3}+y^{2}+z^{2}+6 x y-2 z+1$$

# Решение 

Найдём безусловные (то есть не зависящие от каких-либо ограничений) экстремумы функции

$$
u(x,y,z) \;=\; 3x^3 \;+\; y^2 \;+\; z^2 \;+\; 6xy \;-\; 2z \;+\; 1.
$$

1. Нахождение стационарных точек

Для начала вычислим первые частные производные $u$ по переменным $x$, $y$ и $z$:

\begin{align}

\frac{\partial u}{\partial x} = 9x^2 \;+\; 6y, \quad
\\
\frac{\partial u}{\partial y} = 6x \;+\; 2y, \quad
\\
\frac{\partial u}{\partial z} = 2z \;-\; 2. \quad 
\\

\end{align}

На стационарных точках все эти производные должны быть равны нулю. Следовательно, решаем систему:

\begin{align}

\begin{cases}
9x^2 + 6y = 0, 
\\
6x + 2y = 0, 
\\
2z - 2 = 0.
\end{cases}

\end{align}

Из третьего уравнения непосредственно получаем:

\begin{align}

2z - 2 = 0
\quad\Longrightarrow\quad
z = 1.

\end{align}

Из второго уравнения:

\begin{align}

6x + 2y = 0
\quad\Longrightarrow\quad
3x + y = 0
\quad\Longrightarrow\quad
y = -3x.

\end{align}

Подставим $y = -3x$ в первое уравнение:

\begin{align}

9x^2 + 6(-3x) = 0
\quad\Longrightarrow\quad
9x^2 - 18x = 0
\quad\Longrightarrow\quad
9x(x - 2) = 0.

\end{align}

Отсюда $x = 0$ или $x = 2$. Получаем две стационарные точки:

\begin{align}

(0,\,0,\,1) \quad \text{и} \quad (2,\,-6,\,1).

\end{align}

2. Исследование стационарных точек с помощью матрицы Гессе

Для классификации найдём вторые частные производные (элементы матрицы Гессе):

\begin{align}

H(x,y,z) = 
\begin{pmatrix}
\frac{\partial^2 u}{\partial x^2} & \frac{\partial^2 u}{\partial x \partial y} & \frac{\partial^2 u}{\partial x \partial z}\\[8pt]
\frac{\partial^2 u}{\partial y \partial x} & \frac{\partial^2 u}{\partial y^2} & \frac{\partial^2 u}{\partial y \partial z}\\[8pt]
\frac{\partial^2 u}{\partial z \partial x} & \frac{\partial^2 u}{\partial z \partial y} & \frac{\partial^2 u}{\partial z^2}
\end{pmatrix}
\\[5em] 
H(x,y,z) = 
\begin{pmatrix}
18x & 6 & 0 \\
6 & 2 & 0 \\
0 & 0 & 2
\end{pmatrix}
\quad 

\end{align}

2.1. Точка $(0,0,1)$

Подставим $x=0$, $y=0$, $z=1$ в матрицу Гессе:

\begin{align}

H(0,0,1) \;=\;
\begin{pmatrix}
0 & 6 & 0 \\
6 & 2 & 0 \\
0 & 0 & 2
\end{pmatrix}.

\end{align}

Чтобы понять тип точки, рассмотрим главные миноры:

- Первый главный минор: $\,0$.
- Второй главный минор (детерминант левого верхнего $2\times2$): $0 \cdot 2 - 6 \cdot 6 = -36 < 0$

Отрицательный второй главный минор говорит об *индефинитности* матрицы, следовательно, в точке $(0,0,1)$ функция имеет *седловое положение* (седловую точку).

2.2. Точка $(2,-6,1)$

Подставим $x=2$, $y=-6$, $z=1$ в матрицу Гессе:

\begin{align}

H(2,-6,1) \;=\;
\begin{pmatrix}
36 & 6 & 0 \\
6 & 2 & 0 \\
0 & 0 & 2
\end{pmatrix}

\end{align}

Главные миноры:

- Первый: $36 > 0$.
- Второй (детерминант верхнего левого $2\times2$): $36 \cdot 2 - 6 \cdot 6 = 72 - 36 = 36 > 0$.
- Третий (детерминант всей $3\times3$ матрицы): $36 \times (2 \times 2) - 6 \times \dots$

\begin{align}

\det(H(2,-6,1)) = 36 \cdot (2 \cdot 2) - 6 \cdot \bigl(\det\begin{pmatrix}6 & 0\\ 0 & 2\end{pmatrix}\bigr) 
= 36 \cdot 4 - 6 \cdot (6 \cdot 2) 
= 144 - 72 
= 72 > 0

\end{align}

Поскольку все главные миноры положительны, матрица Гессе *положительно определена*, что означает локальный минимум в точке $$(2,-6,1)$$.

3. Итоговая классификация экстремумов

1. **Точка $(0,0,1)$**: седловая точка (локального минимума или максимума в ней нет).
2. **Точка $(2,-6,1)$**: точка локального минимума.

Проверка значений функции показывает, что при $x \to -\infty$ кубический член $3x^3$ уходит в $-\infty$, а при $x \to +\infty$ — в $+\infty$. Следовательно, функция не ограничена ни сверху, ни снизу, поэтому *глобальных* максимумов и минимумов у неё нет. 

**Ответ:**

- **Локального максимума нет**
- **Локальный минимум**: $(x,y,z) = (2,-6,1)$
- **Седловая точка**: $(x,y,z) = (0,0,1)$

Найдите условный экстремум функции:

$$z=e^{-x-y}$$

Если задано следующее условие:

$$\frac{x^{2}}{4}+y^{2}=5$$

# Решение 

Для решения задачи на поиск условного экстремума функции воспользуемся методом множителей Лагранжа. 

1. Запишем функцию Лагранжа:

   $$L(x,y,\lambda) = e^{-x-y} + \lambda(\frac{x^2}{4} + y^2 - 5)$$

1. Найдем частные производные и приравняем их к нулю:

    \begin{align}

        \frac{\partial L}{\partial x} = -e^{-x-y} + \lambda \frac{x}{2} = 0
        \\
        \frac{\partial L}{\partial y} = -e^{-x-y} + 2\lambda y = 0
        \\
        \frac{\partial L}{\partial \lambda} = \frac{x^2}{4} + y^2 - 5 = 0
   
   \end{align}
   
1. Из первых двух уравнений получаем:

   \begin{align}

      \lambda \frac{x}{2} = \lambda 2y
      \\
      x = 4y

   \end{align}

1. Подставляем это соотношение в уравнение связи:

   \begin{align}

      \frac{(4y)^2}{4} + y^2 = 5
      \\
      4y^2 + y^2 = 5
      \\
      5y^2 = 5
      \\
      y^2 = 1

   \end{align}

1. Отсюда получаем два возможных решения:

   a) $y = 1$, $x = 4$

   b) $y = -1$, $x = -4$

1. Вычислим значения функции в этих точках:

   a) $z(4,1) = e^{-4-1} = e^{-5} = \frac{1}{e^5}$

   b) $z(-4,-1) = e^{4+1} = e^5 $

**Ответ:**

Функция $z=e^{-x-y}$ при условии $\frac{x^{2}}{4}+y^{2}=5$ имеет следующие экстремумы:

1. **Глобальный минимум**: $z_{min} = e^{-5} $ в точке $(x,y) = (4,1)$
1. **Глобальный максимум**: $z_{max} = e^5 $ в точке $(x,y) = (-4,-1)$

# Задание 4

Вам предоставлены данные о людях с наличием или отсутствием заболевания:

Признаки:

* Age — возрастная группа.
* Test — позитивный или негативный тест на заболевание.
* Status — целевая переменная, есть инфекция или нет.

Необходимо реализовать алгоритм Наивного Байеса для решения задачи классификации. Обязательно оцените качество полученного результата по итогу.

```python
# Загрузка данных
data = pd.read_csv("data/asset-v1_SkillFactory+MIFIML-1sem+2024+type@asset+block@sf_exam.csv")
```

```python
data
```

```python
# Реализуем алгоритм Наивного Байеса
class NaiveBayesClassifier:
    """
    Классификатор наивного Байеса.

    Атрибуты:
    priors (dict): Словарь априорных вероятностей классов.
    likelihoods (dict): Словарь условных вероятностей признаков для каждого класса.
    classes (list): Список уникальных классов.
    """

    def __init__(self):
        """
        Инициализирует классификатор наивного Байеса.
        """
        self.priors = {}
        self.likelihoods = {}
        self.classes = []

    def fit(self, X:pd.DataFrame, y:pd.Series) -> None:
        """
        Обучает классификатор наивного Байеса.

        Параметры:
        X: Матрица признаков.
        y: Вектор истинных меток классов.
        """
        self.classes = np.unique(y)
        for cls in self.classes:
            X_cls = X[y == cls]
            self.priors[cls] = len(X_cls) / len(X)
            self.likelihoods[cls] = {}
            for col in X.columns:
                self.likelihoods[cls][col] = X_cls[col].value_counts(normalize=True).to_dict()

    def predict(self, X:pd.DataFrame)->list:
        """
        Предсказывает метки классов для новых данных.

        Параметры:
        X: Матрица признаков новых данных.

        Возвращает:
        Список предсказанных меток классов.
        """
        predictions = []
        for _, x in X.iterrows():
            posteriors = {}
            for cls in self.classes:
                posterior = self.priors[cls]
                for col in X.columns:
                    posterior *= self.likelihoods[cls][col].get(x[col], 0)
                posteriors[cls] = posterior
            predictions.append(max(posteriors, key=posteriors.get))
        return predictions

```

```python
# Разделение данных
X = data[['Test', 'Age_Group']]
y = data['Status']

# Обучение модели
nb = NaiveBayesClassifier()
nb.fit(X, y)

# Предсказания
predictions = nb.predict(X)
```

```python
predictions
```

```python
def accuracy(y_true:list, y_pred:list) -> float:
    """
    Вычисляет точность предсказания модели.

    Параметры:
    y_true: Список истинных значений.
    y_pred: Список предсказанных значений.

    Возвращает:
    Точность предсказания модели.
    """
    if len(y_true) != len(y_pred):
        raise ValueError("Длины списков y_true и y_pred должны совпадать")

    correct_predictions = sum(1 for true, pred in zip(y_true, y_pred) if true == pred)
    total_predictions = len(y_true)

    return correct_predictions / total_predictions


```

```python
def recall(y_true:list, y_pred:list, positive_class:any) -> float:
    """
    Вычисляет метрику recall (полноту) для бинарной классификации.

    Параметры:
    y_true: Список истинных значений.
    y_pred: Список предсказанных значений.
    positive_class: Значение положительного класса.

    Возвращает:
    Значение метрики recall.
    """
    if len(y_true) != len(y_pred):
        raise ValueError("Длины списков y_true и y_pred должны совпадать")

    true_positives = sum(1 for true, pred in zip(y_true, y_pred) if true == pred == positive_class)
    actual_positives = sum(1 for true in y_true if true == positive_class)

    if actual_positives == 0:
        return 0.0  # Избегаем деления на ноль

    return true_positives / actual_positives

```

```python
acc = accuracy(y, predictions)
rec = recall(y, predictions, positive_class="Infected")

print(f"Точность (Accuracy): {acc:.2f}")
print(f"Полнота (Recall) для класса 'Infected': {rec:.2f}")
```

3. Вторая производная $f''(x) = 12x^2 + 18x - 24$

Теперь напишим код для решения задачи

```python
class NewtonOptimizer:
    """
    Оптимизатор методом Ньютона.

    Атрибуты:
    f_prime: Первая производная функции.
    f_double_prime: Вторая производная функции.
    tolerance: Порог сходимости.
    max_iterations: Максимальное количество итераций.
    """

    def __init__(self, f_prime:callable, f_double_prime:callable, tolerance:float, max_iterations:int=100):
        self.f_prime = f_prime
        self.f_double_prime = f_double_prime
        self.tolerance = tolerance
        self.max_iterations = max_iterations

    def optimize(self, x0:float)->float:
        """
        Оптимизирует функцию методом Ньютона.

        Параметры:
        x0: Начальное приближение.

        Возвращает:
        Оптимальное значение.
        """
        x = x0
        for i in range(self.max_iterations):
            fx_prime = self.f_prime(x)
            fx_double_prime = self.f_double_prime(x)
            
            if abs(fx_double_prime) < 1e-10:  # Избегаем деления на очень маленькие числа
                break
            
            x_new = x - fx_prime / fx_double_prime
            
            if abs(x_new - x) < self.tolerance:
                return x_new
            
            x = x_new
        
        return x # Возвращаем последнее значение, если не достигнута сходимость

# Определяем производные функции
def f_prime(x):
    return 4*x**3 + 9*x**2 - 24*x + 7

def f_double_prime(x):
    return 12*x**2 + 18*x - 24

# Создание экземпляра оптимизатора
optimizer = NewtonOptimizer(f_prime, f_double_prime, tolerance=0.0001)

# Оптимизация
x0 = 10
optimal_x = optimizer.optimize(x0)
```

**Ответ:**

```python
print(f"Оптимальное значение x: {optimal_x}")
```

Дано распределение случайной величины, которая отражает вероятность получения разного количества спам-писем в течение дня:


 0 | 1 | 2 | 3  | 4  | 5  | 6 |   7   | 
---|---|---|----|----|----|---|-------|
0.1|0.3|0.1|0.25|0.05|0.05|0.1| **X** |


1. Найдите вероятность получения 7 спам-писем.
2. Найдите математическое ожидание для количества полученных писем.
3. Найдите дисперсию для количества полученных писем. Ответ округлите до сотых.

Задание должно быть решено без использования готовых функций Python.

# Решение

1. Вероятность получения 7 спам-писем

Для начала нам нужно найти значения **X**. Сумма всех вероятностей должна быть равна 1.


```python
# Заданное распределение
distribution = {0: 0.1, 1: 0.3, 2: 0.1, 3: 0.25, 4: 0.05, 5: 0.05, 6: 0.1}

# Вычисление вероятности для 7 спам-писем
p_7 = 1 - sum(distribution.values())

print(f"Вероятность получения 7 спам-писем: {p_7:.2f}")
```

2. Математическое ожидание

```python
# Добавляем вероятность для 7 спам-писем в распределение
distribution[7] = p_7

# Вычисление математического ожидания
expectation = sum(k * p for k, p in distribution.items())

print(f"Математическое ожидание: {expectation:.2f}")
```

3. Дисперсия

```python
variance = sum((k - expectation)**2 * p for k, p in distribution.items())

print(f"Дисперсия: {variance:.2f}")
```

**Ответ:**


```python
print(f"1. Вероятность получения 7 спам-писем: {p_7:.2f}")
print(f"2. Математическое ожидание: {expectation:.2f} спам-писем в день")
print(f"3. Дисперсия (округленная до сотых): {variance:.2f}")
```

# Задание 7

За 8 часов рабочего дня в колл-центр поступает в среднем 16 звонков. Найдите вероятность, что за час в колл-центр поступит не более 4 и не менее 2 звонков. Ответ округлите до тысячных. Задание должно быть решено без использования готовых функций Python.

# Решение 

Для решения данной задачи воспользуемся распределением Пуассона, которое описывает вероятность появления определенного количества событий за фиксированный интервал времени

```python
def factorial(n:int) -> int:
    """
    Вычисляет факториал числа.
    """
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)


def poisson_probability(k:int, lambda_:float) -> float:
   """
   Вычисляет вероятность события в распределении Пуассона.
   """
   return (lambda_**k * math.e**(-lambda_)) / factorial(k)


# Среднее количество звонков за час
lambda_ = 16 / 8  # 16 звонков за 8 часов = 2 звонка в час

# Вычисляем вероятности для k = 2, 3, 4
probability = sum(poisson_probability(k, lambda_) for k in range(2, 5))

# Округляем до тысячных
rounded_probability = round(probability, 3)
```

**Ответ:**

```python
print(f"Вероятность получить 2, 3 или 4 звонка за час: {rounded_probability:.3f}")
```
