## Задание 12

Вероятность, что первый телефон сломан - 0.1. Вероятность того, что второй телефон сломан - 0.3. Найдите вероятность, что оба телефона работают.

**Решение:**

Чтобы найти вероятность того, что оба телефона работают, нам нужно использовать информацию о вероятностях того, что каждый телефон сломан.

1. Вероятности поломки телефонов:
   - Вероятность того, что первый телефон сломан:
   $$ P(A) = 0.1 $$
   - Вероятность того, что второй телефон сломан:
   $$ P(B) = 0.3 $$

2. Вероятности, что телефоны работают:
   - Вероятность того, что первый телефон работает:
   $$ P(A') = 1 - P(A) = 1 - 0.1 = 0.9 $$
   - Вероятность того, что второй телефон работает:
   $$ P(B') = 1 - P(B) = 1 - 0.3 = 0.7 $$

3. Независимость событий:
   Предположим, что события (поломка первого телефона и поломка второго телефона) независимы. Это означает, что состояние одного телефона не влияет на состояние другого. В таком случае вероятность того, что оба телефона работают, можно найти, умножив вероятности того, что каждый телефон работает:
   $$
   P(A' \cap B') = P(A') \times P(B').
   $$

4. Подсчет вероятности:
   Подставим найденные вероятности:
   $$
   P(A' \cap B') = P(A') \times P(B') = 0.9 \times 0.7.
   $$

5. Вычисление:
   $$
   P(A' \cap B') = 0.9 \times 0.7 = 0.63.
   $$

**Ответ:** 0.63

## Задание 15

**Решение:**

Формула:$P(X = k) = C(n, k) \cdot p^k \cdot (1 - p)^{n - k}$, где $n = 6, k = 4, p = 0.5 $. Подставим в формулу и рассчитаем знчение.

$$
C(6, 4) = \frac{6!}{4! \cdot 2!} = 15
$$

$$
P(X = 4) = 15 \cdot (0.5)^4 \cdot (0.5)^2 = 15 \cdot (0.5)^6 = \frac{15}{64}
$$

**Ответ:** $\frac{15}{64}$

**Решение:**

   - $a = 9$ (первая сторона)
   - $b = 5$ (вторая сторона)
   - $h_a = 10$ (высота к стороне a)
   - $h_b$ (высота к стороне b, которую нужно найти)

2. Вспомним формулу площади треугольника:
   S = $\frac{1}{2} \cdot \text{основание} \cdot \text{высота}$

3. Площадь треугольника можно выразить двумя способами:
   S = $\frac{1}{2} \cdot a \cdot h_a = \frac{1}{2} \cdot b \cdot h_b$

4. Подставим известные значения:
   $$\frac{1}{2} \cdot 9 \cdot 10 = \frac{1}{2} \cdot 5 \cdot h_b$$

   $$\frac{1}{2} \cdot 9 \cdot 10 = 45$$

## Задание 19

Найдите объём правильной треугольной пирамиды, стороны основания которой равны 1, а высота равна $\sqrt{3}$

![alt text](https://lms.skillfactory.ru/asset-v1:SkillFactory+MIFIML-1sem+2024+type@asset+block@Специальные_главы_Высшеи__математики_экзамен2.png)

**Решение:**


Для нахождения объёма правильной треугольной пирамиды со стороной основания 1 и высотой $h = \sqrt{3}$, используем формулу для объёма:

$$
V = \frac{1}{3} S_{осн} h
$$

где $S_{осн}$ — площадь основания.

Площадь основания, которое является равносторонним треугольником со стороной $a = 1$, вычисляется по формуле:

$$
S_{осн} = \frac{\sqrt{3}}{4} a^2
$$

Подставим значение $a = 1$:

$$
S_{осн} = \frac{\sqrt{3}}{4} \cdot 1^2 = \frac{\sqrt{3}}{4}
$$

Теперь подставим найденную площадь основания и высоту $h = \sqrt{3}$ в формулу для объёма:

$$
V = \frac{1}{3} \cdot \frac{\sqrt{3}}{4} \cdot \sqrt{3}
$$

Упростим это выражение:

$$
V = \frac{1}{3} \cdot \frac{\sqrt{3} \cdot \sqrt{3}}{4} = \frac{1}{3} \cdot \frac{3}{4} = \frac{1}{4}
$$

Таким образом, объём правильной треугольной пирамиды со стороной основания 1 и высотой $\sqrt{3}$ равен:

$$
V = \frac{1}{4}
$$

**Ответ:** $V = \frac{1}{4}$

## Задание 20

В равнобедренной трапеции основания равны 12 и 27, острый угол равен 60 градусам.

   - Основания: $a = 12, b = 27$

2. Найдем длину боковой стороны c:
   
   В равнобедренной трапеции боковые стороны равны. Из первого поискового результата мы видим, что длина боковой стороны обозначена как c, и она связана с основаниями и углом следующим соотношением:

   $$ \cos^{-1}\left(\frac{15}{2c}\right) = 60° $$

   Преобразуем это уравнение:
   
   $$ \cos(60°) = \frac{15}{2c} $$
   
   $$ \frac{1}{2} = \frac{15}{2c} $$
   
   $$ c = 15 $$

3. Теперь, когда мы знаем длину боковой стороны, можем вычислить периметр:

   Периметр = 2c + a + b
   
   $$ P = 2 \cdot 15 + 12 + 27 = 30 + 39 = 69 $$
