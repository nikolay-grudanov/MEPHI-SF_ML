# Задание 1 (4 балла)
Обратный порядок слов в блоках текста

Файл содержит:

```txt
Привет как дела
На улице идет дождь
Я люблю программирование
```

Критерии оценки:

* Текст разделен на блоки корректно — 2 балла.
* Корректно перевернуты блоки — 2 балла.

````python
def reverse_words_in_blocks(filename: str, block_size: int) -> str:
    """
    Функция разворачивает слова в предложении внутри блоков текста.
    """
    try:
        # Чтение файла
        with open(filename, 'r', encoding='utf-8') as file:
            lines = file.readlines()
        
        # Удаляем лишние символы новой строки
        lines = [line.strip() for line in lines]
        
        # Разделение текста на блоки
        blocks = [lines[i:i + block_size] for i in range(0, len(lines), block_size)]
        
        # Обработка каждого блока
        result = []
        for block in blocks:
            reversed_block = []
            for sentence in block:
                # Разворот слов в предложении
                reversed_sentence = ' '.join(sentence.split()[::-1])
                reversed_block.append(reversed_sentence)
            result.append('\n'.join(reversed_block))
        
        # Объединяем обработанные блоки
        final_result = '\n\n'.join(result)
        
        return final_result
    
    except FileNotFoundError:
        return "Файл не найден."
    except Exception as e:
        return f"Произошла ошибка: {e}"

# Параметры
filename = '1.txt'
block_size = 2  # Количество предложений в блоке

# Вывод результата
output = reverse_words_in_blocks(filename, block_size)
print(output)
````

# Задание 2 (2 балла)

Напишите функцию, которая принимает строку и сжимает её определённым образом. Строки содержат произвольные символы, включая пробелы и спецсимволы, и требуют точного учета длины при кодировании. Сжатие строки происходит сериями одинаковых символов в формате символ + количество, но только если длина сжатой строки не превышает исходную.

Примеры:

* вход: aaabbc → вывод: a3b2c,
* вход: abcd → вывод: abcd (так как сжатая равна).

Критерии оценки:

* Корректно очищен текст — 1 балл.
* Выполняется условие — 1 балл.

````python
def compress_string(s: str) -> str:
    """
    Функция сжимает строку, заменяя повторяющиеся символы на символ и количество повторений.
    """
    # Проверка на пустую строку
    if not s:
        return s

    compressed = []  # Список для хранения сжатой строки
    count = 1  # Счётчик повторений символа

    # Проходим по строке и считаем повторяющиеся символы
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            compressed.append(s[i - 1] + str(count))  # Добавляем символ и его количество
            count = 1  # Сбрасываем счётчик

    # Добавляем последний символ и его количество
    compressed.append(s[-1] + str(count))

    # Преобразуем список в строку
    compressed_str = ''.join(compressed)

    # Возвращаем сжатую строку, если её длина не превышает длину исходной
    return compressed_str if len(compressed_str) <= len(s) else s


# Примеры значений из задания 
s1 = 'aaabbc'
s2 = 'abcd'
# Дополнительный пример для контроля 
s3 = 'aaaabbbcccdde'

print(f'вход: {s1} -> вывод: {compress_string(s1)}\n')
print(f'вход: {s2} -> вывод: {compress_string(s2)}\n')
print(f'вход: {s3} -> вывод: {compress_string(s3)}\n')
````

# Задание 3 (4 балла)

Хаотичные скобки

Реализуйте функцию, которая проверяет, правильно ли расставлены скобки в строке (включая круглые, квадратные и фигурные скобки).

Критерии оценки:

* Функция выполняет условие — 4 балла.

````python
def is_brackets_balanced_ignore_nonbrackets(s):
   """
   Функция проверяет, сбалансированы ли скобки в строке, игнорируя небрекетные символы.

   Args:
       s (str): Исходная строка.

   Returns:
       bool: True, если скобки сбалансированы, иначе False.
   """
   stack = []  # Стек для хранения открывающих скобок
   brackets = {')': '(', ']': '[', '}': '{'}  # Соответствие закрывающих и открывающих скобок

   for char in s:
       if char in brackets.values():  # Если символ — открывающая скобка
           stack.append(char)
       elif char in brackets.keys():  # Если символ — закрывающая скобка
           if stack and stack[-1] == brackets[char]:  # Проверяем соответствие верхнего элемента стека
               stack.pop()  # Удаляем из стека, если скобки совпадают
           else:
               return False  # Если нет соответствия, строка некорректна

   return not stack # Если стек пуст, все скобки сбалансированы

# Примеры для тестирования
examples = [
    "()[]{}",           #True,
    "([{}])",           #True,
    "(]",               #False,
    "([)]",             #False,
    "{[()]}",           #True,
    "{[(])}",           #False,
    "a(b)c[d]",         # True
    "a(b)c[d}",         # False
    "a(b)c[d]e{f}g",    # True
    "(((((((())))))))", # True
    "a(b)c[d]e{f}g}",   # False
    "abc",              # True (нет скобок)
    "",                 # True (нет скобок)
    "(a[b]{c})",        # True (корректная вложенность)
    "(a[b]{c})}",       # False (лишняя закрывающая скобка)
]

# Результаты тестирования
results = {example: is_brackets_balanced_ignore_nonbrackets(example) for example in examples}
print(results)
````

# Задание 4 (4 балла)
Генератор случайных паролей

Напишите функцию, которая генерирует пароль заданной длины. В реализации надо учитывать, что:

* Пароль должен содержать буквы, цифры и специальные символы.
* Длина пароля задается пользователем.

Критерии оценки:

* Функция выполняет все заявленные условия — 4 балла.

````python
# Импорт библиотек
import random
import string

def generate_password(length):
   """
   Функция генерирует пароль заданной длины, содержащий латинские буквы, цифры и специальные символы.

   Args:
       length (int): Длина пароля.

   Returns:
       str: Сгенерированный пароль.
   """
   # Определяем наборы символов
   letters = string.ascii_letters  # Включает латинские буквы верхнего и нижнего регистра
   digits = string.digits          # Включает цифры от 0 до 9
   special_characters = string.punctuation  # Включает специальные символы (!, @, # и т.д.)

   # Если длина меньше 3, пароль будет состоять только из случайных символов,
   # так как мы не соможем включить каждый символ в пароль
   if length <= 3:
       all_characters = letters + digits + special_characters
       return ''.join(random.choices(all_characters, k=length))

   # Гарантируем наличие хотя бы одного символа каждого типа для соблюдения условий задачи
   password = [
       random.choice(letters),
       random.choice(digits),
       random.choice(special_characters)
   ]

   # Заполняем оставшуюся длину случайными символами из всех наборов
   all_characters = letters + digits + special_characters
   password += random.choices(all_characters, k=length - 3)

   # Перемешиваем символы для случайного порядка
   random.shuffle(password)

   return ''.join(password)


# Ввод пользователя с проверкой на корректность (число)
while True:
    user_input = input("Введите длину пароля числом: ")
    if user_input.isdigit():  # Проверяем, что введено число
        password_length = int(user_input)
        break
    else:
        raise ValueError("Введите число, для указания длины пароля.")

# Генерация пароля и вывод результата
print(f"Вы указали длину пароля: {password_length}\nСгенерированный пароль: {generate_password(password_length)}")
````

# Задание 5 (6 баллов)

Эмуляция работы электронной очереди

Напишите класс для симуляции работы электронной очереди, например, в банке.

Критерии оценки:

* Система должна поддерживать добавление клиентов с указанием их приоритета (например, VIP, обычный) — 1 балл.
* Выбор клиента для следующей обработки должен учитывать приоритет — 1 балл.
* Реализуйте отчет для администрации с информацией, сколько времени заняло обслуживание клиентов — 4 балла.


````python
from collections import deque
from time import time, sleep
import random

class ElectronicQueue:
    """
    Класс для электронной очереди.
    """
    def __init__(self):
        self.queue = deque()  # Очередь клиентов (имя и приоритет)
        self.priorities = {'VIP': 0, 'Обычный': 1}  # Приоритеты клиентов
        self.service_times = []  # Список времён обслуживания клиентов (имя, статус, время)

    def add_client(self, name, priority="Обычный"):
        """
        Добавление клиента в очередь.
        :param name: Имя клиента.
        :param priority: Приоритет клиента ("VIP" или "Обычный").
        """
        if priority not in self.priorities:
            raise ValueError(f"Некорректный приоритет: {priority}. Допустимые значения: {list(self.priorities.keys())}")
        
        # Добавляем клиента в очередь
        self.queue.append((name, self.priorities[priority]))
        print(f"Клиент {name} добавлен в очередь с приоритетом {priority}.")

    # Созданим метод для обслуживания клиента для решения задачи это будет заглушка,
    # которая просто делает паузу на 2-5 секунды, потом можно будет переписать под нужный алгоритм
    # Таже сделаем метод приватным, чтобы случайно не вызвать извне.
    def __service_client(self):
        """
        Приватный метод для обслуживания клиента.
        :return: Затраченное время на обслуживание клиента (в секундах).
        """
        service_time = random.uniform(2, 5)  # Генерируем случайное время от 2 до 5 секунд
        sleep(service_time)  # Симуляция обслуживания
        return service_time

    def process_next_client(self):
        """
        Обработка следующего клиента с учётом приоритета.
        :return: Имя обработанного клиента или None, если очередь пуста.
        """
        if not self.queue:
            print("Очередь пуста.")
            return None

        # Находим клиента с наивысшим приоритетом
        highest_priority_index = min(range(len(self.queue)), key=lambda i: self.queue[i][1])
        client_name, client_priority = self.queue[highest_priority_index]

        # Удаляем клиента из очереди
        del self.queue[highest_priority_index]

        # Симулируем время обслуживания через приватный метод
        print(f"Обслуживание клиента {client_name} (Приоритет: {'VIP' if client_priority == 0 else 'Обычный'})...")
        duration = self.__service_client()

        # Логируем время обслуживания
        self.service_times.append((client_name, 'VIP' if client_priority == 0 else 'Обычный', duration))
        
        print(f"Клиент {client_name} обслужен за {duration:.2f} секунд.")
        return client_name

    def generate_report(self):
        """
        Генерация отчёта об обслуживании клиентов.
        :return: Строка с отчётом.
        """
        if not self.service_times:
            return "Клиенты ещё не обслуживались."

        report_lines = ["Отчёт об обслуживании клиентов:"]
        
        total_clients = len(self.service_times)
        
        # Суммируем только время обслуживания (третий элемент каждого кортежа)
        total_time = sum(entry[2] for entry in self.service_times)
        
        report_lines.append(f"Всего клиентов обслужено: {total_clients}")
        report_lines.append(f"Общее время обслуживания: {total_time:.2f} секунд")
        
        for i, (name, status, duration) in enumerate(self.service_times, start=1):
            report_lines.append(
                f"Клиент {i}: Имя — {name}, Статус — {status}, Время обслуживания — {duration:.2f} секунд"
            )
        
        return "\n".join(report_lines)


# Пример использования
if __name__ == "__main__":
    queue = ElectronicQueue()

    # Добавляем клиентов в очередь
    queue.add_client("Иван", priority="Обычный")
    queue.add_client("Мария", priority="VIP")
    queue.add_client("Пётр", priority="Обычный")
    queue.add_client("Анна", priority="VIP")
    queue.add_client("Сергей", priority="Обычный")
    queue.add_client("Екатерина", priority="Обычный")
    queue.add_client("Дмитрий", priority="VIP")

    # Обрабатываем клиентов
    queue.process_next_client()
    queue.process_next_client()
    queue.process_next_client()
    queue.process_next_client()
    queue.process_next_client()
    queue.process_next_client()
    queue.process_next_client()

    # Генерируем отчёт
    print("\n" + queue.generate_report())
````

# Задание 6 (2 балла)

Проверка на «почти палиндром»

Напишите программу, которая проверяет, является ли строка палиндромом или «почти палиндромом». «Почти палиндром» означает, что можно удалить одну букву, чтобы строка стала палиндромом.

Критерии оценки:

* Функция выполняет все заявленные условия — 2 балла.


````python
def is_palindrome(s:str) -> bool:
    """
    Проверяет, является ли строка палиндромом.
    :param s: Строка для проверки.
    :return: True, если строка палиндром, иначе False.
    """
    # Убираем пробелы, приводим к нижнему регистру и удаляем знаки препинания
    s = ''.join(filter(str.isalnum, s)).lower()
    return s == s[::-1]

def is_almost_palindrome(s):
    """
    Проверяет, является ли строка «почти палиндромом».
    :param s: Строка для проверки.
    :return: True, если строка палиндром или почти палиндром, иначе False.
    """
    # Убираем пробелы, приводим к нижнему регистру и удаляем знаки препинания
    s = ''.join(filter(str.isalnum, s)).lower()
    
    if is_palindrome(s):
        return True

    for i in range(len(s)):
        # Удаляем одну букву и проверяем, станет ли строка палиндромом
        modified_s = s[:i] + s[i+1:]
        if is_palindrome(modified_s):
            return True

    return False

# Примеры для проверки
examples = [
    # Проверим известный палиндром, но без учёта регистра и знаков препинания
    "а роза упала на лапу азора",  # Палиндром
    "А роза упала на лапу Азора!", # Палиндром с учётом регистра и знаков препинания
    "абвгд",                       # Не палиндром
    "абцба",                       # Палиндром
    "абцдба",                      # Почти палиндром (удалить 'д')
    # Проверим также и на литинице
    "radar",      # Палиндром
    "racecar",    # Палиндром
    "abca",       # Почти палиндром (удалить 'b')
    "hello",      # Не палиндром и не почти палиндром
]

# Проверка функции
for example in examples:
    result = is_almost_palindrome(example)
    print(f"Строка '{example}' -> {'Палиндром' if result else 'Не палиндром'}")
````

# Задание 7 (9 баллов)

Задача Шредингера
Разработайте программу, которая «стирает» фрагменты текста в файле. Например:

1. Пользователь указывает файл и процент текста, который нужно удалить (например, 30%).
2. Программа случайно выбирает слова или части абзацев и заменяет их на пробел или ..., сохраняя общий объем документа.

Вход:
«Сегодня солнечный день, и я собираюсь гулять в парке с моими друзьями.»

Вывод (удалено ~30%):
«Сегодня ... день, и я собираюсь гулять ... моими друзьями.»

Критерии оценки:

* Чтение файла и ввод параметров пользователя — 3 балла.
* Реализация алгоритма случайного удаления текста — 5 баллов.
* Запись результата и вывод пользователю — 1 балл.

````python
import random

def erase_text(file_path, percentage):
    """
    Удаляет случайные слова из текста в соответствии с указанным процентом.
    :param file_path: Путь к файлу с текстом.
    :param percentage: Процент слов, которые нужно удалить.
    """
    try:
        # Чтение файла
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()
        
        # Разделение текста на слова
        words = text.split()
        total_words = len(words)
        
        # Количество слов для удаления
        words_to_remove = int(total_words * (percentage / 100))
        
        # Выбор случайных индексов для удаления
        indices_to_remove = random.sample(range(total_words), words_to_remove)
        
        # Заменяем выбранные слова на '...'
        for index in indices_to_remove:
            words[index] = '...'
        
        # Объединяем слова обратно в текст
        modified_text = ' '.join(words)
        
        # Сохраняем результат в новый файл
        output_file = "output.txt"
        with open(output_file, 'w', encoding='utf-8') as file:
            file.write(modified_text)
        
        # Вывод результата в консоль
        print("\nРезультат обработки текста:")
        print(modified_text)
        
        print(f"\nОбработанный текст сохранён в файл: {output_file}")
    
    except FileNotFoundError:
        print("Ошибка: Файл не найден.")
    except ValueError:
        print("Ошибка: Указан некорректный процент или файл пуст.")
    except Exception as e:
        print(f"Произошла ошибка: {e}")

# Ввод параметров пользователя
file_path = input("Введите путь к файлу: ")
while True:
    try:
        percentage = float(input("Введите процент текста для удаления (0-100): "))
        if 0 <= percentage <= 100:
            break
        else:
            print("Ошибка: Процент должен быть в диапазоне от 0 до 100.")
    except ValueError:
        print("Ошибка: Введите число.")

# Запуск программы
erase_text(file_path, percentage)


````
